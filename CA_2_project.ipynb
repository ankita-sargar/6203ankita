# **SwarmPlot**
The swarmplot provides a detailed view of the relationship between Age and EmployeeNumber while also considering Menu types through color-coding. Here's the conclusion for this swarmplot:

- The swarmplot offers a fine-grained representation of how employee are distributed within different departments.
- Each data point represents a age and is positioned based on its EmployeeNumber and department.
- The color-coded categories allow for a clear distinction between different department within the plot.
- Observations from the swarmplot include:
  - In some departments, there are employeenumber with age that still achieve  department.
  - The plot provides insights into individual data points, making it useful for identifying outliers and specifice employee with department.

In summary, the swarmplot is a valuable tool for exploring the distribution of emplyeenumber, with the added dimension of departments. It can help identify patterns and outliers that may inform department management decisions.

```python
# Swarm Plot
plt.figure(figsize=(14,6))
sns.swarmplot(x='Age',y="EmployeeNumber", hue='Department', data=data, color='green')
plt.title("Swarm plot")
plt.xlabel("Age")
plt.ylabel("EmployeeNumber")
plt.show()
```


# **Heat map**
A heatmap is a graphical representation of data in a matrix format, where values are represented by different colors. It is particularly useful for visualizing relationships and patterns in complex datasets. In the context of a correlation matrix heatmap:


```python
correlation_matrix = data.corr()
plt.figure(figsize=(22, 10))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Heatmap')

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

# **joint plot**
A joint plot is a type of data visualization used in statistics and data analysis to simultaneously display the relationship between two variables. It combines different types of plots or charts to provide a comprehensive view of the data. Joint plots are especially useful when exploring the association between two continuous variables. There are several common types of joint plots:

```python
sns.jointplot(x='Age', y='EmployeeNumber', data=data, color='purple')
plt.xlabel('Age')
plt.ylabel('EmployeeNumber')
plt.suptitle('Joint Plot: Age vs. EmployeeNumber')
plt.show()
```
